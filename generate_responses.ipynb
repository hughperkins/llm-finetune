````python
import csv
from openai import OpenAI

openai = OpenAI()

prompt_template = """
Here is the backstory for an NPC in an RPG
```
{backstory}
```

Answer the following question or statement to this NPC:
```
{question}
```

Write only the answer to the question or statement. Do not write explanations.
""".strip()

answers = []
qa_pairs = []

for n, question in enumerate(questions):
    print()
    print('question', question)
    prompt = prompt_template.format(backstory=backstory, question=question)
    completion = openai.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "user", "content": prompt}
      ]
    )
    answer = completion.choices[0].message.content
    print('anwer', answer)
    answers.append(answer)
    qa_pairs.append((question, answer))
    # if n >= 3:
    #     break

with open('qa_pairs.csv', 'w') as f:
    dict_writer = csv.DictWriter(f, fieldnames=['n', 'question', 'answer'])
    dict_writer.writeheader()
    for n, qa in enumerate(qa_pairs):
        dict_writer.writerow({'n': n, 'question': qa[0], 'answer': qa[1]})
````
