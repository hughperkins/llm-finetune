```python
import csv
from os.path import expanduser as expand

def save_to_file(data_pairs, filepath):
    with open(filepath, 'w') as f:
        dict_writer = csv.DictWriter(f, fieldnames=['question', 'answer'])
        dict_writer.writeheader()
        for q, a in data_pairs:
            dict_writer.writerow({'question': q, 'answer': a})

save_to_file(train_data, expand('~/Downloads/npc_qa_train.csv'))
save_to_file(test_data, expand('~/Downloads/npc_qa_test.csv'))

!ls -l ~/Downloads/npc_qa*.csv
```
